Adult Dataset

Data Preprocessing Script for Adult Dataset

    Remove invalid or missing lines
    Extracts continuous numerical features (age, fnlwgt, education-num, capital-gain, capital-loss, hours-per-week).
    Encodes the target variable income (>50K as 0, <=50K as 1) and appends it to the features file (adult.txt).
    Encodes the sex attribute (Female as 0, Male as 1) and saves it to a separate labels file (adult_Color.txt).

```python
import os

# Create output directory
os.makedirs('final_data', exist_ok=True)

numerical_lines = []
color_lines = []

# Continuous feature indices
# 0: age, 2: fnlwgt, 4: education-num, 10: capital-gain, 11: capital-loss, 12: hours-per-week
num_indices = [0, 2, 4, 10, 11, 12] 
sex_index = 9
# income_index = 14

# Read dataset
with open('adult_real/adult.data', 'r', encoding='utf-8') as file:
    for line in file:

        parts = [p.strip() for p in line.split(',')]        
        # if len(parts) <= income_index:
        #     continue
        
        if any(not parts[i] or parts[i] == '?' for i in num_indices):
            continue
        sex_val = parts[sex_index]

        # income_val = parts[income_index]        
        # if not sex_val or sex_val == '?' or sex_val.lower() not in ['male', 'female']:
        #     continue
            
        # Encode sex (Female: 0, Male: 1)
        if sex_val.lower() == 'female':
            encoded_sex = '0'
        else:
            encoded_sex = '1'
            
        # # Encode income (>50K: 0, <=50K: 1)
        # if '>50k' in income_val.lower():
        #     encoded_income = '0'
        # elif '<=50k' in income_val.lower():
        #     encoded_income = '1'
        # else:
        #     continue
            
        num_features = [parts[i] for i in num_indices]        
        # num_features.append(encoded_income)
        numerical_lines.append(", ".join(num_features))
        color_lines.append(encoded_sex)

with open('final_data/adult.txt', 'w', encoding='utf-8') as f:
    f.write("\n".join(numerical_lines))

with open('final_data/adult_Color.txt', 'w', encoding='utf-8') as f:
    f.write("\n".join(color_lines))

print("Processing complete.")
print("Features saved to 'final_data/adult.txt'.")
print("Protectd labels saved to 'final_data/adult_Color.txt'.")
print(len(numerical_lines), "valid samples processed.")
```

Bank Dataset
Data Preprocessing Script for Bank Marketing Dataset

    Removes invalid lines or rows with missing/unknown values.
    Extracts continuous numerical features and macroeconomic indicators (age, campaign, pdays, previous, emp.var.rate, cons.price.idx, cons.conf.idx, euribor3m, nr.employed) while deliberately excluding duration to prevent data leakage.
    Encodes the target variable term deposit subscription (yes as 1, no as 0) and appends it to the features file (bank.txt).
    Encodes the marital status attribute (married as 1, single/divorced as 0) and saves it to a separate labels file (bank_Color.txt).

```python

numerical_lines = []
color_lines = []

# Selected continuous features (excluding 'duration' index 10)
# 0: age, 11: campaign, 12: pdays, 13: previous, 15: emp.var.rate, 
# 16: cons.price.idx, 17: cons.conf.idx, 18: euribor3m, 19: nr.employed
num_indices = [0, 11, 12, 13, 15, 16, 17, 18, 19]

marital_index = 2
# target_index = 20

with open('bank_real/bank-additional-full.csv', 'r', encoding='utf-8') as file:
    next(file)
    
    for line in file:
        parts = [p.strip(' "\n') for p in line.split(';')]

        # if len(parts) <= target_index:
        #     continue
            
        if any(not parts[i] or parts[i] == 'unknown' for i in num_indices):
            continue 
        marital_val = parts[marital_index].lower()
        # target_val = parts[target_index].lower()
        
        if marital_val == 'married':
            encoded_marital = '1'
        elif marital_val in ['single', 'divorced']:
            encoded_marital = '0'
        else:
            continue
            
        # if target_val == 'yes':
        #     encoded_target = '1'
        # elif target_val == 'no':
        #     encoded_target = '0'
        # else:
        #     continue
            
        num_features = [parts[i] for i in num_indices]
        
        # num_features.append(encoded_target)
        
        numerical_lines.append(", ".join(num_features))
        color_lines.append(encoded_marital)

with open('final_data/bank.txt', 'w', encoding='utf-8') as f:
    f.write("\n".join(numerical_lines))

with open('final_data/bank_Color.txt', 'w', encoding='utf-8') as f:
    f.write("\n".join(color_lines))

print("Processing complete.")
print("Features saved to 'final_data/bank.txt'.")
print("Marital labels saved to 'final_data/bank_Color.txt'.")
print(len(numerical_lines), "valid samples processed.")
```

Credit card


Data Preprocessing Script for Default of Credit Card Clients Dataset

    Filters out rows containing missing or invalid values.
    Extracts continuous numerical features (LIMIT_BAL, AGE, BILL_AMT1-6, PAY_AMT1-6).
    Encodes the target variable default payment next month (1 for default, 0 for no default) and appends it to the numerical features file (credit_card.txt).
    Encodes the SEX attribute (Male as 1, Female as 0) and saves it to a separate labels file (credit_card_Color.txt).

```python
import pandas as pd

numerical_lines = []
color_lines = []

try:
    df = pd.read_excel('credit_real/default of credit card clients.xls', header=1)
except Exception as e:
    print(f"Error opening the file: {e}")
    exit()

# Numerical columns to extract
num_cols = [
    'LIMIT_BAL', 'AGE', 
    'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6',
    'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6'
]

print("Processing data, please wait...")

for index, row in df.iterrows():
    if row.isnull().all():
        continue
    if row[num_cols].isnull().any() or pd.isna(row['SEX']) or pd.isna(row['default payment next month']):
        continue

    sex_val = str(row['SEX']).strip()
    # target_val = str(row['default payment next month']).strip()
    if sex_val in ['1', '1.0']:
        encoded_sex = '1'
    elif sex_val in ['2', '2.0']:
        encoded_sex = '0'
    else:
        continue
        
    # encoded_target = '1' if float(target_val) == 1 else '0'
    
    num_features = [str(int(row[col])) if float(row[col]).is_integer() else str(row[col]) for col in num_cols]
    
    # num_features.append(encoded_target)
    
    numerical_lines.append(", ".join(num_features))
    color_lines.append(encoded_sex)

with open('final_data/credit_card.txt', 'w', encoding='utf-8') as f:
    f.write("\n".join(numerical_lines))
with open('final_data/credit_card_Color.txt', 'w', encoding='utf-8') as f:
    f.write("\n".join(color_lines))

print("Processing complete!")
print("- Features saved to 'final_data/credit_card.txt'")
print("- Protected labels saved to 'final_data/credit_card_Color.txt'")
print(f"{len(numerical_lines)} valid samples processed.")
```

Census

Reads the raw USCensus1990.data.txt file and dynamically maps column indices from the header row.

    Drops the caseid column to prevent unique identifiers from influencing the model.
    Extracts all 67 remaining numerical/discretized features and saves them to the main features file (uscensus.txt).
    Isolates the iSex attribute and saves it directly to a separate labels file (uscensus_Color.txt) for use in classification or as ground-truth labels for clustering visualization.

```python
import csv

numerical_lines = []
color_lines = []

with open('census_real/USCensus1990.data.txt', 'r', encoding='utf-8') as file:
    reader = csv.reader(file)
    headers = next(reader)
    
    try:
        sex_index = headers.index('iSex')
        caseid_index = headers.index('caseid')
    except ValueError as e:
        print(f"Error: Missing required column in header. {e}")
        exit()
        
    feature_indices = [i for i in range(len(headers)) if i != caseid_index and i != sex_index]

    for parts in reader:
        if not parts or len(parts) != len(headers):
            continue
            
        sex_val = parts[sex_index].strip()
        num_features = [parts[i].strip() for i in feature_indices]
        
        numerical_lines.append(", ".join(num_features))
        color_lines.append(sex_val)

with open('final_data/uscensus.txt', 'w', encoding='utf-8') as f:
    f.write("\n".join(numerical_lines))

with open('final_data/uscensus_Color.txt', 'w', encoding='utf-8') as f:
    f.write("\n".join(color_lines))

print("Processing complete!")
print("- Features saved to 'final_data/uscensus.txt'")
print("- Sex labels saved to 'final_data/uscensus_Color.txt'")
print(f"{len(numerical_lines)} valid samples processed.")
```

**Sampling**

Data Reduction & Stratified Sampling Strategy

    To optimize computational efficiency and model training times, a Stratified Sampling technique was applied across all datasets, retaining exactly 10% of the original data volume.

    This approach was strictly chosen over simple random sampling to ensure Data Integrity and protect against class imbalance. By grouping the data by the label attribute before sampling, we guarantee that the smaller datasets perfectly mirror the statistical distribution and underlying patterns of the original, massive files.

    This results in highly representative subsets that drastically accelerate the algorithm testing phase without compromising the predictive validity of the final machine learning model

```python
print("\n Sampling Adult dataset Keeping 10%")

with open('final_data/adult.txt', 'r', encoding='utf-8') as f:
    features = f.read().splitlines()
with open('final_data/adult_Color.txt', 'r', encoding='utf-8') as f:
    colors = f.read().splitlines()

df = pd.DataFrame({'feature': features, 'color': colors})

original_counts = df['color'].value_counts().to_dict()
print(f"Original size: {len(df)}")
print(f"Original color distribution: {original_counts}")

# sampling(10%)
sampled_df = df.groupby('color', group_keys=False).apply(lambda x: x.sample(frac=0.1, random_state=42))

sampled_counts = sampled_df['color'].value_counts().to_dict()
print(f"Sampled size: {len(sampled_df)}")
print(f"Sampled color distribution: {sampled_counts}")
print(f"Sampled color ratios: {sampled_counts['0']/len(sampled_df)}, {sampled_counts['1']/len(sampled_df)}")
with open('final_data/adult_sampled.txt', 'w', encoding='utf-8') as f:
    f.write("\n".join(sampled_df['feature']))
with open('final_data/adult_Color_sampled.txt', 'w', encoding='utf-8') as f:
    f.write("\n".join(sampled_df['color']))

print("-> Adult sampled successfully!")
```
